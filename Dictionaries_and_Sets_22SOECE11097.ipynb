# ***Program-6 :  Write a Python program that counts the number of occurrences of each character in a given string using a dictionary.***

```python

def count_letters(S):
  d={}
  for i in S:
    if i not in d:
      d[i]=1
    else:
      d[i]+=1
  return(d)
S = input()
d = count_letters(S)
d1 = {}
for i in S:
    try:
        d1[i]+=1
    except KeyError:
        d1[i]=1
if d1 == d:
    print('yes')
else:
    print('no')
```

# ***Program-8 : Create a function that takes a list of dictionaries and sorts them based on a specified key.***

```python
list = [{"name": "Nandini", "age": 20},
	{"name": "Manjeet", "age": 20},
	{"name": "Nikhil", "age": 19}]


print("The list printed sorting by age: ")
print(sorted(list, key=lambda i: i['age']))

print("\r")


print("The list printed sorting by age and name: ")
print(sorted(list, key=lambda i: (i['age'], i['name'])))

print("\r")

print("The list printed sorting by age in descending order: ")
print(sorted(list, key=lambda i: i['age'], reverse=True))
```

# ***Program-9 : Write a program that finds the average value of all the elements in a list of dictionaries.***

```python
list_of_dicts = [
    {'value1': 45, 'value2': 2, 'value3': 53},
    {'value1': 4, 'value2': 5, 'value3': 2},
    {'value1': 1, 'value2': 88, 'value3': 8}
]

total_sum = 0
total_count = 0

for dictionary in list_of_dicts:

    for value in dictionary.values():
        total_sum += value
        total_count += 1

if total_count > 0:
    average = total_sum / total_count
else:
    average = 0


print("Average:", average)
```

# ***Program-10 : Implement a function that takes a list of strings and returns a set of unique characters present in all strings.***

```python
def unique_characters_in_all_strings(string_list):
    common_characters = set(string_list[0])
    for s in string_list[1:]:
        unique_characters = set(s)



        common_characters = common_characters.intersection(unique_characters)

    return common_characters


input_strings = ["nishant", "nick", "their"]
result = unique_characters_in_all_strings(input_strings)
print(result)
```
