# ***Program-7 : Create a function that takes a sentence as input and returns the sentence in reverse order.***

```python
def rev(a):
  e=" "
  b=a.split()
  x=b[::-1]
  c=e.join(x)
  print(c)
rev(input("Enter the sentense:"))
```

# ***Program-9 : Write a function to remove all duplicate characters from a given string.***

```python
a="malangn"
c=list(a)
for i in c:
  b=c.count(i)
  if(b>=2):
    c.remove(i)
    d="".join(c)
    print(d)
```
