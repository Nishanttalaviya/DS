# ***program-9 " Write a program that checks if a given list is sorted in ascending order.***

```python
def is_sorted_ascending(lst):
    for i in range(len(lst) - 1):
        if lst[i] > lst[i + 1]:
            return False
    return True


sorted_list = [1, 2, 3, 4, 5]
unsorted_list = [3, 1, 4, 2, 5]

print(is_sorted_ascending(sorted_list))
print(is_sorted_ascending(unsorted_list))

```
