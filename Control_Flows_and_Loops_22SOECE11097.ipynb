# ***Program-1 : create a program that takes a year as input  and checks if it is a leap year or not***

```python
year = int(input("Enter the year: "))
if(year % 4 == 0):
  print("The year",year," is a leap year.")
else:
  print("The year",year," is not a leap year.")
```

# ***Program-3 : write python program to check if a given number is a prime number.***

```python
num=int(input("enter number:"))
if num>1:
  for i in range(2,num):
    if (num%i)==0:
      print(num,"is not a prime number")
      break
  else:
    print(num,"is a prime number")
else:
  print(num,"is not a prime number")
```

# ***Program-8 : create a loop that print all prime number between 1 and 50.***

```python
prime_nums = []

for num in range (1, 50):
    count = 0
    for i in range(2, (num//2 + 1)):
        if(num % i == 0):
            count = count + 1
            break

    if (count == 0 and num != 1):
        prime_nums.append(num)

print(prime_nums)
```
