# ***Program-1 : Given a list of numbers, create a function to find the sum of all positive numbers.***

```python
def sum():
  a=[1,3,4,-1,-3,-4]
  x=0
  for i in a:
    if(i>0):
      x=x+i
      print(x)
sum()

```

# ***program-3 : Implement a function that returns the factorial of a given number using recursion.***

```python
def factorial(n):
   if n == 1:
       return n
   else:
       return n*factorial(n-1)

num=int(input("Enter the number:"))

if num < 0:
   print("factorial does not exist for negative numbers")
elif num == 0:
   print("The factorial of 0 is 1")
else:
   print("The factorial of", num, "is",factorial(num))
```

# ***program-10 :  Create a function that takes a number as input and prints its multiplication table.***

```python
def table():
  n=int(input("enter the number:"))
  for i in range (1,11):
    print(n,"x",i,"=",n*i)
table()
```
